## ライブラリのインポート

```python
from diffusers import StableDiffusionPipeline

import matplotlib.pyplot as plt
```

# Text to Image

```python
prompt = "A cat in rubber boots riding a bicycle in Ghibli style" #@param {type:"string"}
```

## load model

```python
# load pretrain model
model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", use_auth_token=access_tokens)
model.to("cuda")
```

```python
!mkdir outputs

num = 1

for i in range(num):
  # モデルにpromptを入力し画像生成
  image = model(prompt)["sample"][0]
  # 保存
  image.save(f"outputs/test_{i:04}.png")
```

```python
for i in range(num):
  plt.imshow(plt.imread(f"outputs/test_{i:04}.png"))
  plt.axis('off')
  plt.show()
```
